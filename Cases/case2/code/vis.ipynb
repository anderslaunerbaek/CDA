```python
import numpy as np
from my_class import my_class as my
%matplotlib inline
```

```python
# load
X = np.load("./../data/tmp/X.npy")
Y = np.load("./../data/tmp/Y.npy")
```

```python
# normalize ?
import sklearn.preprocessing as sk_pre
X_norm = sk_pre.normalize(X)
X_norm = X
```

```python
#
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X_norm)
X_pca = pca.transform(X_norm)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=np.argmax(Y, 1), alpha=1.0, label=["1","2"])
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.legend(loc=2)
plt.show()
```
